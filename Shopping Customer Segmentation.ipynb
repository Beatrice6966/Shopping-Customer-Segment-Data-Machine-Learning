```python

import numpy as np

# Example dataset: 100 samples with 2 features
data = np.random.rand(100, 2)  # Replace with your actual data

# Function to calculate inertia for a given number of clusters (k)
def calculate_inertia_for_k(k, data):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    return kmeans.inertia_

# Calculate inertia scores for k values from 1 to 10
inertia_score = [calculate_inertia_for_k(k, data) for k in range(1, 11)]

# Ensure that inertia_score is correctly populated
print(inertia_score)  # Should print a list of 10 values

# Plotting the results
plt.plot(range(1, 11), inertia_score)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia Score')
plt.title('Elbow Method for Optimal k')
plt.show()
```
